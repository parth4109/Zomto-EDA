```python
#here in ourr previous data df country codes were given but i wanted country names
#for that data was in excel sheet 
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

#### merge the two dataframes

```python
df_final = pd.merge(df , df_country ,how = 'left' ,on = "Country Code")
df_final1 = df_final.copy()
df_final.head(2)
```

```python
country_names = df_final["Country"].value_counts().index
country_values = df_final["Country"].value_counts().values
```

```python
country_names
```

```python
country_values
```

```python
plt.pie(country_values[:3],labels = country_names[:3],autopct = "%.2f%%")
```

Observation - Maximum number of records are from India
2nd Max records are from US
3rd Max records are from UK

```python
df_final.columns
```

```python
#this code will make group using the given values
#group_quality = df_final.groupby(['Aggregate rating','Rating color', 'Rating text']).size()
#now to convert it to dataframe we add in the end---
group_quality = df_final.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index()

group_quality
```

```python
group_quality = group_quality.rename(columns = {0:'Rating_count'})
```

```python
group_quality.shape
```

```python
plt.figure(figsize = (12,6))
sns.barplot(data = group_quality,x = 'Aggregate rating',y='Rating_count',hue = 'Rating text',palette = ['brown','red','orange','yellow','green','blue'])
```

Observation2-
as we all know , in real life as well we often see normal distribution aswell
we see that rating is distributed normally with most people giving 3.2 rating out off 5
